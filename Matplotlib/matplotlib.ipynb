### Load necessary libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
```

### Graph

```python
x1= np.random.randint(1,10, size=5)
y1= np.random.randint(1,10, size=5)
```

```python
plt.figure(figsize=(8,5), dpi=100)
# plotting x1 vs y1
plt.plot(x1,y1, label='2x', color='red', linestyle="--", linewidth=2, marker='d', markersize=15, markeredgecolor='blue')
```

```python
x2= np.random.randint(1,10, size=5)
y2= np.random.randint(1,11, size=5)
```

```python
plt.plot(x1,y1, label='2x', color='red', linestyle="--", linewidth=2, marker='d', markersize=15, markeredgecolor='blue')
plt.plot(x2, y2, 'b', label='X^2')

plt.title("Graph by Unique Adhikari", fontdict={'fontname':'Comic Sans MS', 'fontsize': 20})
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.xticks([0,1,2,3,4,])
plt.legend()
plt.savefig("mygraph.png", dpi=300)
plt.plot()
plt.show()
```

### Bar chart

```python
labels= ["A", "B", "C"]
values= [1,4,2]
plt.figure(figsize=(6,4), dpi=100)

bars= plt.bar(labels, values)

patterns= ["/", "O", "*"]
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.title("Bar-chart by Unique Adhikari", fontdict={'fontname':'Comic Sans MS', 'fontsize':20})
plt.xlabel("Alphabets")
plt.ylabel("Frequency")
plt.savefig("mybarchart.png", dpi=300)
plt.show()
```

### Real world examples

Line graph

```python
gas= pd.read_csv("gas_prices.csv")
plt.figure(figsize=(8,5))
plt.title("Gas price over Time (in USD)", fontdict={'fontweight':'bold', 'fontsize':18})

plt.plot(gas.Year, gas.USA, 'b.--', label="United States")
plt.plot(gas.Year, gas.Canada, 'r.-', label="Canada")
plt.plot(gas.Year, gas.Australia, 'g.-', label="Australia")
plt.plot(gas.Year, gas["UK"], 'y.-', label="UK")

plt.xticks(gas.Year[::3].tolist()+[2011])
plt.xlabel("Year")
plt.ylabel("US Dollars")
plt.legend()
plt.savefig("gas_price_figure.png", dpi=300)
plt.show()
```

```python
fifa= pd.read_csv("fifa_data.csv")
fifa.head(0)
```

### Histogram

```python
bins= [40,50,60,70,80,90, 100]
plt.figure(figsize=(9,6))

plt.hist(fifa.Overall, bins=bins, color='#abcdef')

plt.title("Distribution of player skills in FIFA 2018")
plt.xlabel("Skill level")
plt.ylabel("No. of players")

plt.savefig('myhistogram.png', dpi=300)
plt.show()
```

### Piechart

```python
left= fifa.loc[fifa["Preferred Foot"] == "Left"].count()[0]
right= fifa.loc[fifa["Preferred Foot"] == "Right"].count()[0]

plt.figure(figsize=(10,10))

labels= ["Left", "Right"]
colors= ["#abcdef", "#aabbcc"]

plt.pie([left, right], labels= labels, colors=colors, autopct="%.2f %%")
plt.title("Foot preference of FIFA Players")
plt.savefig("mypiechart.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.style.use('ggplot')

fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.2,0,0,.4)

plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')

plt.title('Weight of Professional Soccer Players (lbs)')
plt.savefig("my_explode_piechart", dpi=300)
plt.show()
```

### Box-Whisker chart

```python
plt.figure(figsize=(5,5))
plt.style.use('default')

barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']

bp = plt.boxplot([barcelona, madrid, revs], labels=['FC Barcelona','Real Madrid','NE Revolution'], patch_artist=True, medianprops={'linewidth': 2})

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

for box in bp['boxes']:
    box.set(color='#4286f4', linewidth=2)
    box.set(facecolor = '#e0e0e0' )
    box.set(hatch = '/')
```
